```python
#importing necessary libraries to read, visualize and process the data 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('orders.csv')
df.head(15) 
```

```python
#in shipmode we can see some values as Not available or unknown, we want to treat this as nan
#to convert into null values
df=pd.read_csv('orders.csv',na_values=['Not Available','unknown'])
df.head(15) 
```

```python
df['Ship Mode'].unique()
```

```python
#usually the column name should not have any uppercase and space in between in order to wrok with sql
#rename the columns
#df.rename(columns={'Order Id':'order_id','City':'city'}) #instead of changing each column 
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.columns
```

```python
df.head(15)
```

```python
#since we have discount percent value that tells us that which order will get how much discount. Instead we can make a new column which tels the the price after discount directly
#new derived columns can be: amount that will be discounted [discount_price],how much the company is elling each product[sell_price] and tge profit earned by selling those[profit]
df['discount']=df['discount_percent']*0.01*df['list_price']
df['sell_price']=df['list_price']-df['discount']
df['profit']=df['sell_price']-df['cost_price']
df.head(15)
```

```python
#since cost price, discount percent and list price is no longer needed 
#drop the column
df.drop(['cost_price','list_price','discount_percent'],axis=1,inplace=True)
df.head(15)
```

```python
df.info()
```

```python
#since order date is an object and not in correct datatype
#covert it ot datetime datatype
df['order_date']=pd.to_datetime(df['order_date'])
df.head(15)
```
